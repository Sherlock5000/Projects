# Feature Plots

```python
# Importing required libraries

import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sb
import warnings
warnings.filterwarnings("ignore")
```

```python
# Reading the file containing the data

data = pd.read_csv(r'E:\ESE 588 Pattern Recognition\data.csv')
data.head()
```

```python
# Performing data slicing

data_mod = data[data['Position'] == 'GK']
data_mod1 = data_mod.loc[:, ['Name', 'Position', 'GKDiving', 'GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes', 'Jumping', 'ShotPower', 'Reactions', 'Vision', 'Composure', 'Overall']]
data_mod2 = data_mod1.iloc[0:500,:]
(data_mod2)
```

```python
# Displaying pairplots using seaborn library

sb.pairplot(data_mod2)
```

```python
# Seperation of features

GKDiving = data_mod2.iloc[:, 2].values
GKHandling = data_mod2.iloc[:, 3].values
GKKicking = data_mod2.iloc[:, 4].values
GKPositioning = data_mod2.iloc[:, 5].values
GKReflexes = data_mod2.iloc[:, 6].values
Jumping = data_mod2.iloc[:, 7].values
ShotPower = data_mod2.iloc[:, 8].values
Reactions = data_mod2.iloc[:, 9].values
Vision = data_mod2.iloc[:, 10].values
Composure = data_mod2.iloc[:, 11].values
Overall = data_mod2.iloc[:, -1].values
```

```python
# Plotting probability distribution plot for parameters

sb.distplot(GKDiving)
plt.xlabel('GKDiving')
plt.ylabel('pdf for GKDiving')
plt.show()

sb.distplot(GKHandling)
plt.xlabel('GKHandling')
plt.ylabel('pdf for GKHandling')
plt.show()

sb.distplot(GKKicking)
plt.xlabel('GKKicking')
plt.ylabel('pdf for GKKicking')
plt.show()

sb.distplot(GKPositioning)
plt.xlabel('GKPositioning')
plt.ylabel('pdf for GKPositioning')
plt.show()

sb.distplot(GKReflexes)
plt.xlabel('GKReflexes')
plt.ylabel('pdf for GKReflexes')
plt.show()

sb.distplot(Jumping)
plt.xlabel('Jumping')
plt.ylabel('pdf for Jumping')
plt.show()

sb.distplot(ShotPower)
plt.xlabel('ShotPower')
plt.ylabel('pdf for ShotPower')
plt.show()

sb.distplot(Reactions)
plt.xlabel('Reactions')
plt.ylabel('pdf for Reactions')
plt.show()

sb.distplot(Vision)
plt.xlabel('Vision')
plt.ylabel('pdf for Vision')
plt.show()

sb.distplot(Composure)
plt.xlabel('Composure')
plt.ylabel('pdf for Composure')
plt.show()

sb.distplot(Overall)
plt.xlabel('Overall')
plt.ylabel('pdf for Overall')
plt.show()
```

```python
# Importing libraries for performing linear regression and splitting data into training and testing sets

from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression as lr
lin_reg = lr()
```

```python
GKDiving_train, GKDiving_test, Overall_train, Overall_test = tts(GKDiving, Overall, train_size = 4/5, test_size = 1/5)
lin_reg.fit(GKDiving_train.reshape(-1, 1), Overall_train.reshape(-1, 1))
Overall_predict = lin_reg.predict(GKDiving_test.reshape(-1, 1))

plt.scatter(GKDiving_train.reshape(-1, 1), Overall_train.reshape(-1, 1), color = 'greenyellow')
plt.plot(GKDiving_train.reshape(-1, 1), lin_reg.predict(GKDiving_train.reshape(-1, 1)))
plt.show()
```
